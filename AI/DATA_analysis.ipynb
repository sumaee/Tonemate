```python
import os
import pandas as pd

path = r'C:\Users\SSAFY\Desktop\특화프로젝트\train'
columns = ['나이', '스타일', '성별', '장르', '식별자_wav']

data = []
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith('.wav'):
            file_name = os.path.splitext(file)[0]
            file_info = file_name.split('_')
            index = f"{file_info[0]}_{file_info[1]}"
            features = file_info[2:]
            data.append([index] + features)

df = pd.DataFrame(data, columns=['가수식별'] + columns)
df = df.drop('식별자_wav', axis=1)
df = df.drop_duplicates()
#df.set_index('식별_숫자', inplace=True)
```
